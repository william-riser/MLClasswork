```python
from scipy.integrate import quad
from numpy import exp
from numpy import inf

def integrand(x):
    return 2*exp(-2 * x)

(result, error) = quad(integrand, 2, inf)

print('Probability: ', result)
print('Number of people to wait longer than 2 seconds: %d' % (result * 1000000000))
```

```python

SAT = genfromtxt('sat.csv', delimiter=',')

n = len(SAT)
avg = np.mean(SAT)
std = np.std(SAT)

print('Mean: ', avg)
print('Standard deviation: ', std)
```

```python
import matplotlib.pyplot as plt
plt.hist(SAT, bins=20, density=True, alpha=0.6)

x = np.linspace(avg-4*std, avg+4*std, 100)
y = (1 / (std * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - avg) / std) ** 2)
plt.plot(x, y, color='red', linewidth=2)

plt.xlabel('Value')
plt.ylabel('Probability Density')

plt.show()
```
