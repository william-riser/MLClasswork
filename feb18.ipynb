```python
import numpy as np
from numpy import genfromtxt
from numpy import exp
from scipy.integrate import quad
```

```python
X = genfromtxt('lunch_wait_time.csv', delimiter=',')
avg = np.mean(X)
theta = 1/avg
```

```python
def p(x):
    return theta * exp(-theta*x)
def f(x):
    return x * p(x)
```

```python
mu, err = quad(f, 0, np.inf)
print('Integration Vs Avg: %.3f, %.3f'%(mu,avg))
```

```python
def f(x):
    return (2*x + 1)*p(x)
avg = np.mean(2*X + 1)

mu, err = quad(f, 0, np.inf)
print('Integration Vs Avg: %.3f, %.3f'%(mu,avg))
```

### Practice Sampling w/ Python

```python
import matplotlib.pyplot as plt

samples = np.random.exponential(scale=3, size=50000)
print(np.mean(samples))

theta = 1/3
x = np.linspace(0, 15, 100)
y = theta * exp(-theta*x)
plt.plot(x, y)

plt.hist(samples, bins=50, density=True)
plt.show()
```

```python

samples = np.random.rand(50000)
print('Prob less than 0.3: ', np.sum(samples < 0.3)/50000)
print('Prob less than 0.6: ', np.sum(samples < 0.6)/50000)
```

### Practice Inversion Sampling

```python
mu = 2
theta = 1/mu
n = 50000
v = np.random.rand(n)
samples = -np.log(1 - v)/theta

x = np.linspace(0, 15, 100)
y = theta * exp(-theta*x)
plt.plot(x, y, 'red')
plt.hist(samples, bins=50, density=True)
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
print('Mean: %.3f'%np.mean(samples))
```
