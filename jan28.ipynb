```python
# Height distribution of male and female

avg_male = 70
avg_female = 64.5
sd_male = 3
sd_female = 2.5

x = np.linspace(55, 85, 1000)
x2 = np.linspace(55, 77, 1000)

pdf = stats.norm.pdf(x, avg_male, sd_male)
pdf2 = stats.norm.pdf(x2, avg_female, sd_female)

plt.figure(figsize=(10, 6))
plt.plot(x, pdf, color='blue', label='Gaussian PDF')
plt.plot(x2, pdf2, color='red')

x_fill = np.linspace(50, 67, 1000)
pdf_fill = stats.norm.pdf(x_fill, avg_female, sd_female)
plt.fill_between(x_fill, pdf_fill, color='lightgrey', alpha=0.5)
```
